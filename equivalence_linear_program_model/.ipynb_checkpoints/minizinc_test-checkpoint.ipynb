```python
import numpy as np
import random
import scipy as sc
import gurobipy as gp
import copy
import pandas as pd


def transf_mps_in_std_form():

        #set_inst = "miplib/collection"
        #set_inst_test = "collection_npz/"
        #inst = "2club200v15p5scn"
        #inst = "8div-n59k10"
        #inst = "23588"
        #extention = ".mps"
        ########################################################
        set_inst = "lp"
        set_inst_test = "lp_npz/"
        inst = "25fv47"
        extention =".txt"
        inst = "emps"
        #extention =".exe"
        #extention =".f"
        
        # read the mps file using gurobi package
        model = gp.read("data_set/"+set_inst+"/"+inst+extention)

        # for an efficient manipulation of sparsity, 
        # use of lil_matrix format
        # get the constraint matrix in sparse csr_matrix format
        A = model.getA()

        print(A.shape)

        # get the right-hand side (RHS) values in sparse csr_matrix format
        rhs = model.getAttr("RHS", model.getConstrs())

        # get the objective coefficients in sparse csr_matrix format
        obj = model.getAttr("Obj", model.getVars())

        # Get the constraints senses (Sense) in sparse csr_matrix format
        senses = model.getAttr("Sense", model.getConstrs())
        #print("len_senses =", len(senses))
        #print("len_rhs =", len(rhs))
        #print("obj =", obj)
        #print("len_obj =", len(obj))
        #print("senses =", senses)

        # get coefs ctrs with the '>' sense and transform them in the '<' sense
        for ctr_rank in range(len(senses)):
            if senses[ctr_rank] == '>' or senses[ctr_rank] == '>=':
               print("salut")
               A[ctr_rank, :] = A[ctr_rank, :].tolil().multiply(-1).tocsr()
               rhs[ctr_rank] = -1 * rhs[ctr_rank]
        
        # get the standard form by concatenation with hstack and vstack
        rhs = sc.sparse.csr_matrix([0] + rhs).reshape((len(rhs)+1, 1))
        obj = sc.sparse.csr_matrix(obj)
        return sc.sparse.hstack([sc.sparse.vstack([obj, A],"csr"),rhs],"csr").astype(np.int64)
    
# This function generate a permutation matrice of sixe lines X lines
def gen_permut_matrix(not_self_lines):

        # Create an identity matrix of size lines x lines
        identity_matrix = np.eye(not_self_lines)

        # Generate a random permutation of integes from 0 to lines-1
        while True: 
            permutation_lines = np.random.permutation(not_self_lines)

            # Permute the rows of the identity matrix according to the
            # random permutation_lines
            identity_matrix = identity_matrix[permutation_lines]
            permut = random.randint(0,not_self_lines-1)
            if identity_matrix[permut, permut] != 1:
               return sc.sparse.csr_matrix(identity_matrix,dtype=int)
    

 
# This function generate two equal or not equal matrices by permutation of lines
# and columns. The matrices are same sized
def generate_two_matrices_by_permut(A,proportional,inst,set_inst,set_inst_test,equal,type):
    
    #A is transformation in a std form of an LP. So we update the attributs
    # shape accordingly
    scale_A = False
    lines, columns = A.shape

    # ceate a mask B of the matrix A: 
    B = copy.deepcopy(A)
        
    if proportional:
       # convert to lil_matrix for efficient manipulation of rows
       A = A.tolil()
       B = B.tolil()
       # instancuate the recorder of scaling
       list_coef_B = []
       # generate scaled matrices A and B
       for i in range(lines):
            coef_B = random.randint(0,10)
            B[i,:] = copy.deepcopy(A[i,:].multiply(coef_B))
            list_coef_B = list_coef_B + [i,coef_B]
       # reconvert to csr_matrix for efficient operation on sparses matrices
       A = A.tocsr()
       B = B.tocsr()    
    

    # Create a permutation matrix of size lines x lines
    permut_matrix_lines = gen_permut_matrix(lines)
    #print("Permut_lines=",permut_matrix_lines)

    # Create a permutation matrix of size columns x columns
    permut_matrix_columns = gen_permut_matrix(columns)
    #print("Permut_cols=",permut_matrix_columns)

    # Permute the lines and columns of A to obtain B and ensure that the 
    # coeffient of the resulting matrices multiplication are integers
    B = (permut_matrix_lines @ B @ permut_matrix_columns).astype(np.int64)
    
               

    # get the number of times a test matrices example has been generated
    directory = "data_set/miplib/"+set_inst+"/"+set_inst_test
    file = directory+set_inst+"_scaled_matrices/"+inst
    file0 = directory+set_inst+"_scaled_matrices/"
    num = np.load(file0+"/rank_mat_"+type+".npy")[0] + 1
    #num = 0

    # save all traces for generating the instances test matrices
    save_testing_matrices_in_npz(file,num,A,B,"_A","_B",type)
    save_testing_matrices_in_npz(file,num,permut_matrix_lines,
                                          permut_matrix_columns,
                                          "_PermutLin",
                                          "_PermutCol",type)
        

    if not equal:
        # permute some positions of coefficients of a lines and columns of 
        # the matrix B in this way

        # record it
        type = "not_eq_pro"
        file = directory+set_inst+"not_equal_matrices/"
        
        # Generate a random size of the submatrix of B to permut it lines and its columns
        first_rand_index_line = random.randint(0, lines //2)
        num_of_lines = random.randint(0, lines //2)
        first_rand_index_column = random.randint(0, columns//2)
        num_of_columns = random.randint(0, columns//2)
    
        # Create an permutation matrix of size num_of_lines x num_of_lines
        permut_coefs_line = self.gen_permut_matrix(num_of_lines)

        # Create an permutation matrix of size num_of_columns x num_of_columns
        permut_coefs_col = self.gen_permut_matrix(num_of_columns)

        # Permute the submatrix of B according to the permut_coefs matrices
        B[first_rand_index_line:first_rand_index_line + num_of_lines,
               first_rand_index_column:first_rand_index_column + num_of_columns] = (permut_coefs_line 
               @ B[first_rand_index_line:first_rand_index_line + num_of_lines,
               first_rand_index_column:first_rand_index_column + num_of_columns] 
               @ permut_coefs_col).astype(np.int64)
           
        # save all traces for generating the instances test matrices
        save_testing_matrices_in_npz(file,num,permut_coefs_line,
                                          permut_coefs_col,
                                          "_NotPermutLin",
                                          "_NotPermutCol",type)
        # update the numbber of times a test matrices example has been generated
        np.save(file0+"/rank_mat_NotEqual"+type+".npy",np.array([num,
                                                                     first_rand_index_line,
                                                                     num_of_lines,
                                                                     first_rand_index_column,
                                                                     num_of_columns]))
    else:
        # update the numbber of times a test matrices example has been generated
        np.save(file0+"/rank_mat_"+type+".npy",np.array([num]+list_coef_B))
        
        

    return A,B
    
# save the generated matrices of test for permutation equality or
# proportionality
def save_testing_matrices_in_npz(directory,num,A,B,namA,namB,type):
        
        
        
    # save the generated matrices with new name using 'num' to differ from
    # the old example
        
    rank_type = str(num)+"_"+type
    sc.sparse.save_npz(directory+namA+rank_type+".npz",A)
    sc.sparse.save_npz(directory+namB+rank_type+".npz",B)
        

"""
equal = True
proportional = True
set_inst = "collection"
set_inst_test = "collection_npz/"
type = "eq_pro"

df = pd.read_csv('data_set/miplib/metadata_collection_set.csv')
name_column = df['InstanceInst.']
for instance in name_column[1:]:
    inst = instance
    # get the std form A of a MIP of MIPLIB
    A = transf_mps_in_std_form()

    # Generate the second matrix B to test 
    A,B = generate_two_matrices_by_permut(A,proportional,inst,set_inst,set_inst_test,equal,type)
    
    
#print(name_column)"""
```

```python


def i_is_mobile_elt(permut, i):

    if len(permut) > 1:
       if i == 0:
          if  permut[0][1] > 0 and permut[0][0] > permut[1][0]:
              return True
          else :
              return False
       
       elif i <= len(permut) - 2:
           
         if permut[i][1] < 0 and permut[i][0] > permut[i-1][0]:
              return True
         if permut[i][1] > 0 and permut[i][0] > permut[i+1][0]:
              return True
         else :
             return False
       else :
          if permut[-1][1] < 0 and permut[-1][0] > permut[-2][0]:
             return True
          else :
             return False
    else :
        return False

def mobile_elt_exist(permut):

    for i in range(len(permut)):
        if i_is_mobile_elt(permut, i) :
            return True
    return False

def largest_mobile_elt(permut):
    
    largest_mobile = -1
    ind_largest_mobile = -1
    for i in range(len(permut)):
        if i_is_mobile_elt(permut, i) :
            if permut[i][0] > largest_mobile :
                largest_mobile = permut[i][0]
                ind_largest_mobile = i
    return ind_largest_mobile,largest_mobile
        
def swap_with_adjac_k_pointo_elt(k,permut):

    if permut[k][1] < 0 :
       keep = permut[k-1]
       permut[k-1] = permut[k]
       permut[k] = keep
    else :
       keep = permut[k+1]
       permut[k+1] = permut[k]
       permut[k] = keep

def reverse_dir_all_larg_than_k(largest_mobile_elt,permut):

    for elt in permut:
        if elt[0] > largest_mobile_elt :
            if elt[1] > 0 :
                elt[1] = -1
            else :
                 elt[1] = 1
            


# algorithm for generating permutations
def JohnsonTrotter(n):
    
    # initialize the first permutation
    permut = [[i,-1] for i in range(n)]
    while mobile_elt_exist(permut):
          k, largest_mobile_elt_ = largest_mobile_elt(permut)
          swap_with_adjac_k_pointo_elt(k,permut)
          reverse_dir_all_larg_than_k(largest_mobile_elt_,permut)
          permutation = [elt[0] for elt in permut]
          #print(permut)
          print(permutation)

JohnsonTrotter(4)
```

```python
from multiprocessing import Process, Queue

def func_prog(queue, data):
        result = data * 2
        queue.put(result)

def timer(timeout, data, func_prog):
    if __name__ == "__main__":
        q = Queue()
        p = Process(target=func_prog, args=(q, data))
        p.start()
        p.join(timeout)  # Wait for the process to finish
        returned_value = q.get()
        print(f"Returned value: {returned_value}")
```
