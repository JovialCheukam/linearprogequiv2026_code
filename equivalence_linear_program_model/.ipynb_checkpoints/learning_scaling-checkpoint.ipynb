```python

def may_be_fast_find_non_duplicate(row):
    non_duplicate = []
    l = len(row)
    sort_row = sorted(row)
    find_new_duplicate = False
    for i in range(l-1):
        if sort_row[i] != sort_row[i+1]:
            if not find_new_duplicate:
                non_duplicate.append(sort_row[i])
                
            if i == l-2:
                    non_duplicate.append(sort_row[i+1]) 
            find_new_duplicate = False 
        else :
            find_new_duplicate = True
    return non_duplicate

def find_non_duplicate(row):
    temoin_list = {}
    non_duplicate = []
    for i in row:
        temoin_list[i] = 0
    for i in row:
        temoin_list[i] = temoin_list[i] + 1
    for key in temoin_list.keys():
        if temoin_list[key] == 1:
            non_duplicate.append(key)
    return non_duplicate

def get_permutation(row1,row2):
    l = len(row1)
    assert(l == len(row2))
    non_duplicates = find_non_duplicate(row1)
    permut_vect = [-1]*l
    for i in range(l):
        for non_duplicate_j in non_duplicates:
            if row1[i] == non_duplicate_j:
                for k in range(l):
                    if row1[i] == row2[k]:
                       permut_vect[i] = k
                       break
    return permut_vect

def is_same_permut_for_non_diplicates_elts(row1,row2_corr1,row3,row4_corr3):
    permut_vect1 = get_permutation(row1,row2_corr1)
    permut_vect2 = get_permutation(row3,row4_corr3)
    l = len(permut_vect1)
    for i in range(l):
        if permut_vect1[i] != -1 and permut_vect1[i] != permut_vect2[i]:
            return False
    return True 


def learn_wrong_scaling(A,B,proport_coefs,learned_wrong_scaling):
    num_lines = len(proport_coefs)
    row1 = A[proport_coefs[0][2],:].multiply(proport_coefs[0][1])
    row2_corr1 = B[proport_coefs[0][3],:].multiply(proport_coefs[0][0])
    for i in range(1,num_lines):
        row3 = A[proport_coefs[i][2],:].multiply(proport_coefs[i][1])
        row4_corr3 = B[proport_coefs[i][3],:].multiply(proport_coefs[i][0])
        if not is_same_permut_for_non_diplicates_elts(row1,row2_corr1,row3,row4_corr3):
            if (proport_coefs[0][3],proport_coefs[0][3]) not in learned_wrong_scaling.keys():
                learned_wrong_scaling[(proport_coefs[0][3],proport_coefs[0][3])]= [(proport_coefs[i][3],proport_coefs[i][3])]
            else : 
                learned_wrong_scaling[(proport_coefs[0][3],proport_coefs[0][3])].append((proport_coefs[i][3],proport_coefs[i][3]))
            return False
    return True

```

```python
import scipy as sc

file = 'data_set/netlib/lp/lp_npz/lp_equal_matrices/sc50b'
namA = '_A59_'
namB = '_B59_'
rank_type = 'eq'


A = sc.sparse.load_npz(file+namA+rank_type+".npz")
B = sc.sparse.load_npz(file+namB+rank_type+".npz")

def write_matrix(file,A,A_str,num_line, num_col):

    file.write(A_str + ' = [|')
    for i in range(num_line):
        for j in range(num_col):
            file.write(str(A[i,j]) + ', ')
        if i == num_line-1 and j == num_col-1:
           file.write('|];\n\n')
        else :
            file.write('|\n')

def write_dzn_format(A,B):
    num_line, num_col = A.shape
    
    with open('A_B_eq.dzn', "w") as file:
        file.write('\n')
        file.write('m = '+ str(num_line) + ';\n')
        file.write('n = '+ str(num_col) + ';\n\n')
        file.write('min_k = '+ str(1) + ';\n')
        file.write('max_k = '+ str(10) + ';\n\n')
        file.write('test_scale_rows_only = '+ str(0) + ';\n\n')
        write_matrix(file,A,'A',num_line, num_col)
        write_matrix(file,B,'B',num_line, num_col)

write_dzn_format(A,B)
```

```python
import minizinc
from minizinc import Status
import numpy as np
from datetime import timedelta

file = 'data_set/netlib/lp/lp_npz/lp_equal_matrices/afiro'
namA = '_A59_'
namB = '_B59_'
rank_type = 'eq'


A = sc.sparse.load_npz(file+namA+rank_type+".npz")
B = sc.sparse.load_npz(file+namB+rank_type+".npz")

model = """
      % Verify that matrices A and B are the same up to rows and columns 
      % permutations with scaling on individual rows.
      % Unfeasible will be returned if there exist no vP and vQ matrices that
      % can transform A in to B.
      % In this example, A and B are int, so is vX. 

      int: n;
      int: m;
      int: min_k;
      int: max_k;
      int: test_scale_rows_only;

      % First and second matrices
      array[1..m,1..n] of int: A;
      array[1..m,1..n] of int: B;

      % Permutation matrix for the rows
      array[1..m,1..m] of var {0,1}: vP;
                   
      % Permutation matrix for the columns
      array[1..n,1..n] of var {0,1}: vQ;

      % Intermediate result after a permutation of the rows
      array[1..m,1..n] of var int: vX;

      % Scaling of the rows
      array[1..m] of var min_k..max_k: vS1;
      array[1..m] of var min_k..max_k: vS2;
      var bool: rows_only;
                 
      % Permutation of the rows
      constraint forall(i in 1..m, j in 1..n)(
      sum(k in 1..m)(vP[i,k] * A[k,j]) = vX[i,j]
      );

      % Permutation of the columns with scaling
      constraint bool2int(rows_only) = test_scale_rows_only;
      constraint if not rows_only then forall(i in 1..m, j in 1..n)(
      sum(k in 1..n)(vX[i,k] * vQ[k,j] * vS1[i]) = B[i,j] * vS2[i]
      ) else forall(i in 1..m, j in 1..n)(vX[i,j] * vS1[i] = B[i,j] * vS2[i]) endif;

      % The rows and columns of the row-permutation matrix should sum to 1
      constraint forall(j in 1..m)(
      sum(i in 1..m)(vP[i,j]) = 1
      );
      constraint forall(i in 1..m)(
      sum(j in 1..m)(vP[i,j]) = 1
      );

      % The rows and columns of the column-permutation matrix should sum to 1
      constraint if not rows_only then forall(j in 1..n)(
      sum(i in 1..n)(vQ[i,j]) = 1
      ) else true endif ;
      constraint if not rows_only then forall(i in 1..n)(
      sum(j in 1..n)(vQ[i,j]) = 1
      ) else true endif ;
  
      % The scaling should be higher than 0.0
      constraint forall(i in 1..m)(vS1[i] >= 0);
      constraint forall(i in 1..m)(vS2[i] >= 0);

      solve satisfy;
      """

Model = minizinc.Model()
Model.add_string(model)
solver = minizinc.Solver.lookup("chuffed")
inst = minizinc.Instance(solver, Model)

# get the shape of the matrices to test
(m,n) = A.shape

# connect the data instances
inst["n"] = n
inst["m"] = m
inst["min_k"] = 1
inst["max_k"] = 10
result = inst.solve(time_limit=timedelta(minutes=10), free_search=True)
print(result.solution)
```
