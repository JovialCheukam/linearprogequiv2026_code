```python
# This code test if two matrices are equal by swapping lines and columns

import numpy as np
import time
import random


class Algo_test_equal_permut_mat:

    def __init__(self, A, B):

        assert(type(A) == np.ndarray and type(B) == np.ndarray)
        assert(A.dtype == "int64" and A.dtype == "int64")
        self.A = A
        self.B = B

    # This function check if two sets (set_1, set_2) have the same elements 
    # in nlogn elementary operations.  set_1 and set_2 should be numpy vectors
    # with same length
    def check_equal_sets(self,set_1,set_2): # Complexity of O(nlogn)

        # set_1 and set_2 should be numpy vectors of integers
        assert(type(set_1) == np.ndarray and type(set_2) == np.ndarray)
        assert(set_1.dtype == "int64" and set_2.dtype == "int64")

        # get the shape of set_1
        shape_set = set_1.shape 

        # set_1 and set_2 should have the same length
        assert(shape_set == set_2.shape)

        # set_1 and set_2 are first sorted:
        sort_set_1 = np.sort(set_1)     # Complexity of O(nlogn)
        sort_set_2 = np.sort(set_2)

        # test if the two sets are equal:
        for i in range(shape_set[0]):
            if sort_set_1[i] != sort_set_2[i]:
                return False
        return True



    # This function test if the set of coefficients of each line of the matrix A is equal
    # to the set of coefficients of a line of the matrix B or if the set of coefficients 
    # of each column of the matrix A is equal to the set of coefficients of a column of 
    # the matrix B

    def check_equal_lines_or_columns(self,line_or_column,A,B):

        # A and B should be numpy ndarray of integers
        assert(type(self.A) == np.ndarray and type(self.B) == np.ndarray)
        assert(self.A.dtype == "int64" and self.A.dtype == "int64")

        # get the shape of set_1
        shape_mat = self.A.shape 

        # set_1 and set_2 should have the same length
        assert(shape_mat == self.B.shape)

        # get if the test is on lines or on columns:
        if line_or_column == 'lines' :
            N = shape_mat[0]
        else :
            N = shape_mat[1]

        # instanciate the recorder of matching same set in lines or columns:
        match_tab = [0]*N
        

        # instanciate the index of lines or columns to check in B:
        index_tabs_to_check_B = [1]*N
        

        # test if the set of coefficients of each lines of A is the set of coefficients 
        # of a line of B or test if the set of coefficients of each column of A is the 
        # set of coefficients of a line of B

        # initiate the check of a line or a column
        a_tab_is_already_tested = False

        for i_l_or_c in range(N):   # complexity of O(n^3logn)
            if a_tab_is_already_tested :
                if match_tab[i_l_or_c-1] == 0:
                    print("false0",i_l_or_c-1,line_or_column)
                    return False

            # search the set of line  A in lines of B or
            # search the set of column  A in colums of B 
            for j_l_or_c in range(N):  # complexity of O(n^2logn)
                if index_tabs_to_check_B[j_l_or_c] == 1:
                    if line_or_column == 'lines' :
                        test_passed = self.check_equal_sets(self.A[i_l_or_c,:],self.B[j_l_or_c,:]) # complexity of O(nlogn)
                    else :
                        test_passed = self.check_equal_sets(self.A[:, i_l_or_c],self.B[:,j_l_or_c]) # complexity of O(nlogn)
                    if test_passed :
                       match_tab[i_l_or_c] = 1
                       index_tabs_to_check_B[j_l_or_c] = 0
                       break
            a_tab_is_already_tested = True
                    
        if match_tab[N - 1] == 0:
                   print("false1",line_or_column)
                   return False
        return True
                        
                    
    # This function test if one matrix A is obtained by swapping the lines and columns 
    # of the matrix B. The algorithm just check that the elements of the lines of one
    # matrix are the elements of lines of the other matrix. It also do the same for the
    # columns.

    def check_equal_matrices_by_permutation(self):

        # A and B should be numpy ndarray of integers
        assert(type(self.A) == np.ndarray and type(self.B) == np.ndarray)
        assert(self.A.dtype == "int64" and self.A.dtype == "int64")

        # get the shape of set_1
        shape_mat = self.A.shape 

        # set_1 and set_2 should have the same length
        assert(shape_mat == self.B.shape)

        
        # test if the set of coefficients of each lines of A is the set of coefficients 
        # of a line of B or test if the set of coefficients of each column of A is the 
        # set of coefficients of a line of B

        test_passed = self.check_equal_lines_or_columns("lines",self.A,self.B)
        if not test_passed :
            return False
        else:
            return self.check_equal_lines_or_columns("columns",self.A,self.B)
```

```python

# This class generate a testor if two matrices are proportional
# by  swapping lines and columns. It inherits from the class
# Algo_test_equal_permut_mat

from bisect import bisect_left

class Algo_test_proport_permut_mat(Algo_test_equal_permut_mat):

    # This function test if the two sets set_1 and set_2 are proportional 
    # It return the coefficient of proportionality k if it is the case. 
    # Typically set_1 and set_2 are lines of two matrices
    ##################### Fixed bugs ########################
    # checking when there are 0s in the sets to be checked proportional
    def check_proportional_sets_lines(self, set_1, set_2):
        
        # set_1 and set_2 should be numpy vectors of integers
        assert(type(set_1) == np.ndarray and type(set_2) == np.ndarray)
        assert(set_1.dtype == "int64" and set_2.dtype == "int64")

        # get the shape of set_1
        shape_set = set_1.shape 

        # set_1 and set_2 should have the same length
        assert(shape_set == set_2.shape)

        # set_1 and set_2 are first sorted:
        sort_set_1 = np.sort(set_1)     # Complexity of O(nlogn)
        sort_set_2 = np.sort(set_2)
        
        for i in range(1,shape_set[0]):
            if sort_set_1[i] * sort_set_2[0] != sort_set_2[i] * sort_set_1[0] :
                return (False, (False,False))
        # print("ici=",set_1,set_2,sort_set_1[0],sort_set_2[0])
        return (True, (sort_set_1[0],sort_set_2[0]))


    # Searches for x in the sorted list by using bisect_left of module bisect.
    # Returns True if found.
    def binary_search(self, x, tab):
    
        pos = bisect_left(tab, x)
        if pos != len(tab) and tab[pos] == x:
           return True
        return False



    # This function test if there is |set_1| couple of numbers such that for each
    # couple of two numbers, one in the set set_1 and another in the set set_2 are 
    # proportional while the coefficient of proportionality is a ration a/b where (a,b) 
    # is an element of the set tabCoef. It return False if not. Typically, set_1 and set_2
    # are columns of two matrices and sort_set_2 is the sorted version of set_2
    def check_proportional_sets_columns(self, tabCoef, set_1, sort_set_2):
        
        # set_1 and set_2 should be numpy vectors of integers
        assert(type(set_1) == np.ndarray and type(sort_set_2) == np.ndarray)
        assert(set_1.dtype == "int64" and sort_set_2.dtype == "int64")

        # get the shape of set_1 and tabCoef
        shape_set = set_1.shape 
        num_coefs = len(tabCoef)

        # set_1 and set_2 should have the same length
        assert(shape_set == sort_set_2.shape and shape_set[0] == num_coefs)

        # instanciate the indexes the recorder if an element of set_1 was found 
        # proportional to an element of set_2
        set_1_is_proport = np.zeros((shape_set[0],shape_set[0]))

        i = 0
        while i < shape_set[0] :  # complexity of O(n^3)
            for j in range(num_coefs):
                if set_1_is_proport[i,j] == 0 :
                    (a, b) = tabCoef[j]
                    tab =  a * sort_set_2
                    # print("bin test", "(i,j)=",i, j,"(a,b)=",a,b,"set_1[i]=",set_1[i], "b*set_1[i]=",b*set_1[i], "dans tab=",tab)
                    test_passed = self.binary_search(b*set_1[i], tab)
                    if test_passed :    # When the jth coef matches for proportionality
                        set_1_is_proport[i,j] = 1               # record that matching between element i of set_1 and coef j
                        for other_i in range(shape_set[0]):     # Ensure that all other element other_i of set_1
                           if set_1_is_proport[other_i,j] == 0: # element other_i of set_1 should 
                               set_1_is_proport[other_i,j] = 2  # not be tested to match with coef j
                        break                                   # Stop the search of a matching elt i of set_1 with coef j 
                    else:                 # Each time the elt i do not match with coef j, record it also                    
                        set_1_is_proport[i,j] = -1 

            if not test_passed :             # When an elt of set_1 didn't find its matching coef  
               if i == 0 :                   # If it was the first elt of set_1
                  return False               # No need to continue the search       
               else :                        # If not, the i - 1 th elt of set_1 found it matching coef before
                   i = i - 1                 # Search another matching coef of the i-1 elt of set_1
                   for j in range(shape_set[0]):        # Before that search for the previous  
                       if set_1_is_proport[i,j] == 1:   # matching coef j for the i-1 elt of set_1
                          set_1_is_proport[i + 1,j] = 0 # Set the coef j to be matchable with other elts of set_1
                          break
            else :
                i = i + 1      # Search a matching coef for the next elt i of set_1

        return True


    # This function find the sequence of transpositions of a permutation existing
    # between two vectors with the same set of components values. It return the
    # vector v where the ith component is transposed to position v[i]
    def get_permut(self, vect_1, vect_2):
        
        # get the shape of set_1
        shape_line = vect_1.shape 

        # set_1 and set_2 should have the same length
        assert(shape_line == vect_2.shape)

        # gen the mask of the
        permut = [0]*shape_line[0]

        # initiate the transpose position recorder
        trans_pos = [0]*shape_line[0]
        
        for i in range(shape_line[0]):
            for j in range(shape_line[0]):
                if trans_pos[j] == 0:
                   if vect_1[i] == vect_2[j]:
                      permut[j]=i
                      trans_pos[j] = 1
                      break
        return permut

    # This function test if each line of the matrix A is proportional at 
    # one line of matrix B or if for each column of the matrix A there 
    # is a column of the matrix B such that each element of the column 
    # of A is proportional to an element of the column B while the 
    # coefficient of proportionality is a ration a/b where the couple (a,b)
    # is an element of the set tabCoef. It return False if not
    ##################### Fixed bugs ########################
    # backtracking when a first tab of A match with a tab in B and a second
    # tab of A don't match a tab in B
    def check_proportional(self,line_or_column,tabCoef,A,B):

        # A and B should be numpy ndarray of integers
        assert(type(self.A) == np.ndarray and type(self.B) == np.ndarray)
        assert(self.A.dtype == "int64" and self.A.dtype == "int64")

        # get the shape of set_1
        shape_mat = self.A.shape 

        # set_1 and set_2 should have the same length
        assert(shape_mat == self.B.shape)

        # get if the test is on lines or on columns:
        if line_or_column == 'lines' :
                N = shape_mat[0]
        else :
                N = shape_mat[1]

        # instanciate the recorder of the coefs of proportionality
        proport_coefs = [0]*N

        # test if the set of coefficients of each lines of A is the set of coefficients 
        # of a line of B or test if the set of coefficients of each column of A is the 
        # set of coefficients of a line of B

        # instanciate the index of lines or columns of A to check in B:
        match_A_B = np.zeros((N,N))

        # initiate the the first tab of A to search for its matching in B
        i_l_or_c = 0
        while i_l_or_c < N :                          # complexity of O(n^3logn) or O(n^5)

            for j_l_or_c in range(N):                 # Search the tab of B to match with tab i_l_or_c of A
                 
                if match_A_B[i_l_or_c,j_l_or_c] == 0: # If tab j_l_or_c of B isn't assign to match with 
                                                      # tab i_l_or_c of A
                   if line_or_column == "lines":
                       (test_passed, proport_coefs[i_l_or_c]) = self.check_proportional_sets_lines(
                            self.A[i_l_or_c,:],self.B[j_l_or_c,:]) # complexity of O(nlogn)
                       
                   else :
                       # print("A_=", self.A[:,i_l_or_c])
                       # print("B_=", self.B[:,j_l_or_c])
                       sortB = np.sort(self.B[:,j_l_or_c])
                       test_passed = self.check_proportional_sets_columns(
                            tabCoef, self.A[:, i_l_or_c],sortB) # complexity of O(n^3)
                   if test_passed :          # When the tab j_l_or_c of B is found to match tab i_l_or_c of A
                       match_A_B[i_l_or_c,j_l_or_c] = 1 # record that matching between tabs i_l_or_c and j_l_or_c
                       for i in range(N):
                           if match_A_B[i,j_l_or_c] == 0:
                               match_A_B[i,j_l_or_c] = 2 # all other tab i of A should not be tested
                                                         # to match with tab j_l_or_c of B
                       break                             # Stop the search of a matching tab j_l_or_c for B for 
                                                         # tab i_l_or_c of A
                   else :                    # tab j_l_or_c of B tested but don't match with i_l_or_c of A
                       match_A_B[i_l_or_c,j_l_or_c] = -1
            
            if not test_passed :             # When a tab of A didn't find its match in B  
               if i_l_or_c == 0 :            # If it was the first tab of A
                  return (False,False)
               else :                        # The i_l_or_c - 1 th tab of A found it match in B before
                   i_l_or_c = i_l_or_c - 1   # Search another matching tab in B for i_l_or_c - 1 th tab of A
                   for j in range(N):
                       if match_A_B[i_l_or_c,j] == 1: # Find the last matching tab j in B for i_l_or_c - 1 tab of A
                          match_A_B[i_l_or_c + 1,j] = 0 # Set the tab j in B to be matchable with other tab of A
                          break
            else :
                i_l_or_c = i_l_or_c + 1      # Search a matching tab in B for the next i_l_or_c tab of A

        #if line_or_column == "lines":
            # print(line_or_column,"match_A_B=",match_A_B)
        return (True, proport_coefs)    


   # This function test if each line of the matrix A is proportional at 
    # one line of matrix B or if for each column of the matrix A there 
    # is a column of the matrix B such that each element of the column 
    # of A is proportional to an element of the column B while the 
    # coefficient of proportionality is a ration a/b where the couple (a,b)
    # is an element of the set tabCoef. It return False if not
    ##################### Fixed bugs ########################
    # backtracking when a first tab of A match with a tab in B and a second
    # tab of A don't match a tab in B
    ##############################################################
    # This function is faster than the check_proportional() function because
    # it checks the columns proportionality in O(nlogn) operations while the
    # check_proportional() function checks the columns proportionality in O(n^3) 
    # operations. This gain in time efficiency is due to the transformation of
    # matrices A and B as equal by permutations of lines and columns using the 
    # cross products property of proportions 
    def fast_check_proportional(self,line_or_column,tabCoef,A,B):

        # A and B should be numpy ndarray of integers
        assert(type(self.A) == np.ndarray and type(self.B) == np.ndarray)
        assert(self.A.dtype == "int64" and self.A.dtype == "int64")

        # get the shape of set_1
        shape_mat = self.A.shape 

        # set_1 and set_2 should have the same length
        assert(shape_mat == self.B.shape)

        # get if the test is on lines or on columns:
        if line_or_column == 'lines' :
                N = shape_mat[0]
        else :
                N = shape_mat[1]
                # normalize the matrices A and B as equal by permutations of
                # lines and columns using the cross products property of proportions
                for [k_1, k_2, l_1, l_2] in tabCoef.values():
                    A[l_1,:] = k_2 * A[l_1,:]
                    B[l_2,:] = k_1 * B[l_2,:]

        # instanciate the recorder of the coefs of proportionality
        proport_coefs = {}
        for i in range(N):
            proport_coefs[i] = [0,0,0,0]

        # test if the set of coefficients of each lines of A is the set of coefficients 
        # of a line of B or test if the set of coefficients of each column of A is the 
        # set of coefficients of a line of B

        # instanciate the index of lines or columns of A to check in B:
        match_A_B = np.zeros((N,N))

        # initiate the the first tab of A to search for its matching in B
        i_l_or_c = 0
        while i_l_or_c < N :                          # complexity of O(n^3logn)

            for j_l_or_c in range(N):                 # Search the tab of B to match with tab i_l_or_c of A
                 
                if match_A_B[i_l_or_c,j_l_or_c] == 0: # If tab j_l_or_c of B isn't assign to match with 
                                                      # tab i_l_or_c of A
                   if line_or_column == "lines":
                       (test_passed, Coefs) = self.check_proportional_sets_lines(
                            self.A[i_l_or_c,:],self.B[j_l_or_c,:]) # complexity of O(nlogn)
                       if test_passed :
                          proport_coefs[i_l_or_c][0] = Coefs[0]
                          proport_coefs[i_l_or_c][1] = Coefs[1]
                          proport_coefs[i_l_or_c][2] = i_l_or_c
                          proport_coefs[i_l_or_c][3] = j_l_or_c
                       
                   else :
                       # print("A_=", self.A[:,i_l_or_c])
                       # print("B_=", self.B[:,j_l_or_c])
                       test_passed = self.check_equal_sets(self.A[:, i_l_or_c],self.B[:,j_l_or_c]) # complexity of O(nlogn)
                   if test_passed :          # When the tab j_l_or_c of B is found to match tab i_l_or_c of A
                       match_A_B[i_l_or_c,j_l_or_c] = 1 # record that matching between tabs i_l_or_c and j_l_or_c
                       for i in range(N):
                           if match_A_B[i,j_l_or_c] == 0:
                               match_A_B[i,j_l_or_c] = 2 # all other tab i of A should not be tested
                                                         # to match with tab j_l_or_c of B
                       break                             # Stop the search of a matching tab j_l_or_c for B for 
                                                         # tab i_l_or_c of A
                   else :                    # tab j_l_or_c of B tested but don't match with i_l_or_c of A
                       match_A_B[i_l_or_c,j_l_or_c] = -1
            
            if not test_passed :             # When a tab of A didn't find its match in B  
               if i_l_or_c == 0 :            # If it was the first tab of A
                  return (False,False)
               else :                        # The i_l_or_c - 1 th tab of A found it match in B before
                   i_l_or_c = i_l_or_c - 1   # Search another matching tab in B for i_l_or_c - 1 th tab of A
                   for j in range(N):
                       if match_A_B[i_l_or_c,j] == 1: # Find the last matching tab j in B for i_l_or_c - 1 tab of A
                          match_A_B[i_l_or_c + 1,j] = 0 # Set the tab j in B to be matchable with other tab of A
                          break
            else :
                i_l_or_c = i_l_or_c + 1      # Search a matching tab in B for the next i_l_or_c tab of A

        # if line_or_column == "lines":
            # print(line_or_column,"Coefs=",proport_coefs)
        return (True, proport_coefs)    


        

    # This function test if the matrix A is proportional to the matrix B 
    # by permutation of lines and columns
    def check_proportional_matrices_by_permutation(self):

        # A and B should be numpy ndarray of integers
        assert(type(self.A) == np.ndarray and type(self.B) == np.ndarray)
        assert(self.A.dtype == "int64" and self.A.dtype == "int64")

        # get the shape of set_1
        shape_mat = self.A.shape 

        # set_1 and set_2 should have the same length
        assert(shape_mat == self.B.shape)

        
        # test if each line of the matrix A is proportional at one line of matrix B 
        # then if for each column of the matrix A there is a column of the matrix B 
        # such that each element of the column of A is proportional to an element 
        # of the column B while the coefficient of proportionality is a ration a/b 
        # where the couple (a,b) is an element of the set tabCoef. 

        (test_passed, tabCoef) = self.check_proportional("lines",0, self.A,self.B)
        # print("tabCoef =", tabCoef)
        if not test_passed :
            print("Test failed on lines")
            return False
        else:
            print("Test on columns")
            (test_passed,no) = self.check_proportional("columns", tabCoef, self.A,self.B)
            return test_passed


    # This function test if the matrix A is proportional to the matrix B 
    # by permutation of lines and columns. It uses the fast_check_proportional() 
    # function instead of the check_proportional() function. So it is faster than
    # than the check_proportional_matrices_by_permutation() function
    
    def fast_check_proportional_matrices_by_permutation(self):

        # A and B should be numpy ndarray of integers
        assert(type(self.A) == np.ndarray and type(self.B) == np.ndarray)
        assert(self.A.dtype == "int64" and self.A.dtype == "int64")

        # get the shape of set_1
        shape_mat = self.A.shape 

        # set_1 and set_2 should have the same length
        assert(shape_mat == self.B.shape)

        
        # test if each line of the matrix A is proportional at one line of matrix B 
        # then if for each column of the matrix A there is a column of the matrix B 
        # such that each element of the column of A is proportional to an element 
        # of the column B while the coefficient of proportionality is a ration a/b 
        # where the couple (a,b) is an element of the set tabCoef. 

        (test_passed, tabCoef) = self.fast_check_proportional("lines",0, self.A,self.B)
        # print("tabCoef =", tabCoef)
        if not test_passed :
            print("Test failed on lines")
            return False
        else:
            print("Test on columns")
            (test_passed,no) = self.fast_check_proportional("columns", tabCoef, self.A,self.B)
            return test_passed
 
```

```python
# Fourth check of the algorithm code: with two randomn matrices lines X columns which 
# are equal or not by permutation of lines and columns

class MatForTestPermut:

    def __init__(self, lines = None, 
                       columns = None, 
                       min_val = None, 
                       max_val = None, 
                       equal = None, 
                       proportional = None):

        self.lines = lines
        self.columns = columns
        self.min_val = min_val
        self.max_val = max_val
        self.equal = equal
        self.proportional = proportional

    # This function generate a permutation matrice of sixe lines X lines
    def gen_permut_matrix(self,not_self_lines):

        # Create an identity matrix of size lines x lines
        identity_matrix = np.eye(not_self_lines)

        # Generate a random permutation of integes from 0 to lines-1
        permutation_lines = np.random.permutation(not_self_lines)

        # Permute the rows of the identity matrix according to the random permutation_lines
        return identity_matrix[permutation_lines]
    

 

    # This function generate two equal or not equal matrices by permutation of lines
    # and columns. The matrices are same sized
    def generate_two_matrices_by_permut(self):

        # Generate a matrix lines X columns with random integers
        A = np.random.randint(self.min_val, self.max_val + 1, size=(self.lines, self.columns))
        # print("Init_A=",A)

        #Ceate a mask B of the matrix A: 
        B = np.zeros((self.lines, self.columns))

        if self.proportional:
            for i in range(self.lines):
                coef_A = random.randint(1,10)
                coef_B = random.randint(1,10)
                B[i,:] = coef_B*A[i,:]
                A[i,:] = coef_A*A[i,:]
                
                # print("coefs=", coef_A, coef_B)

        # Create a permutation matrix of size lines x lines
        permut_matrix_lines = self.gen_permut_matrix(self.lines)
        # print("Permut_lines=",permut_matrix_lines)

        # Create a permutation matrix of size columns x columns
        permut_matrix_columns = self.gen_permut_matrix(self.columns)
        # print("Permut_cols=",permut_matrix_columns)

        # Permute the lines and columns of A to obtain B and ensure that the 
        # coeffient of the resulting matrices multiplication are integers
        if not self.proportional:
            B = A
        B = (permut_matrix_lines @ B @ permut_matrix_columns).astype(np.int64)

        

        if not self.equal:
           # permute some positions of coefficients of a lines and columns of 
           # the matrix B in this way
       
           # Generate a random size of the submatrix of B to permut it lines and its columns
           first_rand_index_line = random.randint(0, self.lines //2)
           num_of_lines = random.randint(0, self.lines //2)
           first_rand_index_column = random.randint(0, self.columns//2)
           num_of_columns = random.randint(0, self.columns//2)
    
           # Create an permutation matrix of size num_of_lines x num_of_lines
           permut_coefs_line = self.gen_permut_matrix(num_of_lines)

           # Create an permutation matrix of size num_of_columns x num_of_columns
           permut_coefs_col = self.gen_permut_matrix(num_of_columns)

           # Permute the submatrix of B according to the permut_coefs matrices
           B[first_rand_index_line:first_rand_index_line + num_of_lines,
               first_rand_index_column:first_rand_index_column + num_of_columns] = (permut_coefs_line 
               @ B[first_rand_index_line:first_rand_index_line + num_of_lines,
               first_rand_index_column:first_rand_index_column + num_of_columns] 
               @ permut_coefs_col).astype(np.int64)
    
        return A,B

    def save_testing_matrices_in_npy(self,A,B):
        
        np.save("matrix_A.npy",A)

        np.save("matrix_B.npy",B)

    def get_testing_matrices_from_npy(self):

        A = np.load("matrix_A.npy")

        B = np.load("matrix_B.npy")
        return A, B

```

```python
# Execution of the test of matrices and the algorithm for equality by permutation 
# of lines an columns

lines   = 1000    # Number of lines in the matrix
columns = 2000  # Number of columns in the matrix
min_val = 1     # Minimum value for the random integers 
max_val = 1300   # Maximum value for the random integers
equal = True
proportional = False

# Initiate the matrices generator class for testing
testing = MatForTestPermut(lines, columns, min_val, max_val, equal,proportional)

# Generate two matrices to test 
A,B = testing.generate_two_matrices_by_permut()

# Initiate the testor 
testor = Algo_test_equal_permut_mat(A,B)

# initiate the time compter
start_time = time.time()

# execute the test of equality of matrices by permutation
result = testor.check_equal_matrices_by_permutation()

# record the time of execution
end_time = time.time()

# compute the execution time 
execution_time = end_time - start_time

# print the results
print("The matrices are equal by permutation: ",result)
print(f"Execution time: {execution_time} seconds")
```
