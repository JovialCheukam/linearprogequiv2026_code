```python
import pandas as pd

df = pd.read_csv("results_exp/results_and_remaining_of_experiments.csv")
i = 0
for model in df['model']:
    if model == 'algo_test'or model == 'enhanced_test':
        df.loc[i,'solver'] = model
    i = i + 1
```

```python
preambule1 = "\\documentclass{book}\n\\usepackage[landscape,a3paper,left=1in,right=1in,top=1in,"
preambule2 = "bottom=1in]{geometry}\n\\usepackage{amsmath}\n\\usepackage{xcolor}\n\\usepackage{tikz}\n"
preambule3 = "\\usepackage{pgfplots}\n\\usepackage{booktabs}\n\\usepackage{multirow}\n\\usepackage{amsmath,"
preambule4 = "amssymb}\n\n\\begin{document}\n\n\\vspace{2cm}\n\n"
preambule5 = "\\begin{figure}\\pgfplotsset{small,samples=10}\n\\centering\n\\begin{tabular}{lr}"

end_graphic = "\n\\end{tabular}\\\\\n\\ref{named}\\end{figure}\n\n"


def first_head_env_graphi(title,type_axis,xlabel,ylabel,list_of_graphs_legend):
    return "\\begin{tikzpicture}\n\\begin{"+type_axis+"}[scaled ticks=false, legend columns = -1,legend entries=" + list_of_graphs_legend+", legend to name = named,\ntitle={" + title + "},xlabel=" + xlabel +",ylabel=" + ylabel+"]\n"

def seconds_head_env_graphi(title,type_axis,xlabel,ylabel,p):
    if 2 == p:
        ylabel_written = ",ylabel=" + ylabel
    else:
        ylabel_written =""
        
    return "\\begin{tikzpicture}\n\\begin{"+type_axis+"}[scaled ticks=false,\ntitle={" + title + "},xlabel=" + xlabel+ylabel_written+"]\n"

def add_function(function,min_val,max_val):
    return "\\addplot[domain="+min_val+":"+max_val+"] {"+ function + "};\n"
    
def add_plot(color,mark,coordinates):
    return "\\addplot[color="+color+",mark="+mark+"] coordinates {"+ coordinates + "};\n"   

def stats(type_inst,model,results_test,solver):
    char = len(df[(df['solver'] == solver) & (df['type of instance'] == type_inst) & (df['result of the test'] == results_test) & (df['model'] == model)])
    return str(char)

def end_env_graphic(type_axis):
    return "\\end{"+type_axis+"}\n\\end{tikzpicture}\n"

models = [('permutation_model','black','oplus'),('quadratic_model','blue','otimes'),('theorem_model','red','o'),('algo_test','teal','*'),('enhanced_test','brown','diamond')]
models_strings = ['\\mbox{  }permutation\\_model\\mbox{  }','\\mbox{  }quadratic\\_model\\mbox{  }','\\mbox{  }theorem\\_model\\mbox{  }','\\mbox{  }algo\\_test\\mbox{  }','\\mbox{  }enhanced\\_test']
solvers = ['coinbc','cp-sat','chuffed','gecode','algo_test','enhanced_test','sicstus']
type_inst_list = ['equal by permutation','not equal by permutation']
results_test_list = ['equivalent','not equivalent','unknown due to timeout']
ylabel = "Number of solved instances"
xlabel = "Number of time out"

min_val = "0"
max_val = "70"
function = "x"
bissector = add_function(function,min_val,max_val)

type_inst_list = [('not equal by permutation',['unknown due to timeout','not equivalent'],
                   'non-equivalent instances'),
                  ('equal by permutation and scaling',['unknown due to timeout','equivalent'],
                   'equivalent scaled instances'),
                  ('not equal by permutation and scaling',['unknown due to timeout','not equivalent'],
                   'non-equivalent scaled instances')]

def write_legend(name_plots):
    plots = ['bissector\\mbox{  }']+name_plots
    i = 0
    legend ="{"
    for name in plots:
        if i == 0:
           legend = legend + name
        elif i == len(plots)-1:
            legend = legend + ";," + name + "}"
        else:
            legend = legend + ";," + name
        i = i + 1
    return legend
    
    
def graph_coordinates_models(type_inst,results_test_list,models,solvers):
        
        for (model,color,mark) in models:
            coordinates = ""
            for solver in solvers:
                x = stats(type_inst,model,results_test_list[0],solver)
                y = stats(type_inst,model,results_test_list[1],solver)
                if not (x == '0' and y == '0'):
                   coordinates = coordinates + "("+ x + "," + y +")\n"
            coordonate = add_plot(color,mark,coordinates)
            file.write(coordonate)


def graphic_model(type_inst,results_test_list,type_name,p):
        title = "Methods for solved "+type_name
        graphic_head = seconds_head_env_graphi(title,type_axis,xlabel,ylabel,p)
        file.write(graphic_head)
        file.write(bissector)
        graph_coordinates_models(type_inst,results_test_list,models,solvers)
        end_env = end_env_graphic(type_axis)
        file.write(end_env)

with open("graphics/graphics.tex", "w") as file: 
    file.write(preambule1)
    file.write(preambule2)
    file.write(preambule3)
    file.write(preambule4)
    file.write(preambule5)
    list_of_graphs_legend = write_legend(models_strings)
    title = "Methods for solved equivalent instances"
    type_axis = "axis"
    first_head = first_head_env_graphi(title,type_axis,xlabel,ylabel,list_of_graphs_legend)
    file.write(first_head)
    bissector = add_function(function,min_val,max_val)
    file.write(bissector)
    graph_coordinates_models('equal by permutation',['unknown due to timeout','equivalent'],models,solvers)
    end_env = end_env_graphic(type_axis)
    file.write(end_env+"&")
    i = 1
    for (type_inst,results_test_list,type_name) in type_inst_list:
        graphic_model(type_inst,results_test_list,type_name,i)
        if i == 1 :
            file.write("\\\\\n")
        if i == 2 :
            file.write(" & ")
        i = i + 1
    file.write('\n')
    file.write(end_graphic)
    file.write("\\end{document}")
```

```python
import datetime


def stats_solvers(type_inst,name_inst,solver):
    df_time = df[(df['instance name'] == name_inst) & (df['solver'] == solver) & (df['type of instance'] == type_inst)]
    format_seconds = "%H:%M:%S"
    time_second = 0
    for time_string in df_time['execution time']:
        date_time = datetime.datetime.strptime(time_string, format_seconds)
        time_delta = datetime.timedelta(hours=date_time.hour,minutes=date_time.minute,seconds=date_time.second)
        time_second = time_second + time_delta.total_seconds()
    #print(df_time)
    if df_time.empty :
        #print("sover =",solver)
        return False
    else :
        ind = df_time['Unnamed: 0'].tolist()[0]
        #print(ind)
        #if solver =='sicstus':
           #print(solver)
        size = df_time.loc[ind,'number of variables'] + df_time.loc[ind,'number of constraints']
        return (size,time_second,solver)

type_inst = 'equal by permutation'
#model = 'permutation_model'
solver = 'chuffed'
name_inst = '25fv47'
#stats_solvers(type_inst,name_inst,solver)

solvers_ = [('coinbc','black','+'),('cp-sat','blue','x'),('chuffed','teal','square'),('gecode','red','triangle'),('algo_test','orange','star'),('enhanced_test','green','diamond'),('sicstus','brown','*')]
solvers_strings = ['\\mbox{ }coinbc\\mbox{ }','\\mbox{ }cp-sat\\mbox{ }','\\mbox{ }chuffed\\mbox{ }','\\mbox{ }gecode\\mbox{ }','\\mbox{ }algo\\_test\\mbox{ }','\\mbox{ }enhanced\\_test\\mbox{ }','\\mbox{ }sicstus\\mbox{ }']
xlabel = "Number of variables + constraints"
ylabel = "Total time on all models"
type_axis = 'loglogaxis'
min_val = "50"
max_val = "10020"
function = "x"
bissector = add_function(function,min_val,max_val)

name_inst_list = ["25fv47","80bau3b","adlittle","afiro","agg","agg2","agg3","bandm","beaconfd","blend",
              "bnl1","bnl2","boeing1","boeing2","bore3d","brandy","capri","cycle","czprob",
              "d2q06c","d6cube","degen2","degen3","dfl001","e226","etamacro","fffff800","finnis",
              "fit1d","fit1p","fit2d","fit2p","forplan","ganges","gfrd-pnc","greenbea","greenbeb",
              "grow15","grow22","grow7","israel","kb2","lotfi","maros","maros-r7","modszk1",
              "nesm","perold","pilot","pilot.ja","pilot.we","pilot4","pilot87","pilotnov","recipe",
              "sc105","sc205","sc50a","sc50b","scagr25","scagr7","scfxm1","scfxm2","scfxm3",
              "scorpion","scrs8","scsd1","scsd6","scsd8","sctap1","sctap2","sctap3","seba","share1b",
              "share2b","shell","ship04l","ship04s","ship08l","ship08s","ship12l","ship12s","sierra",
              "stair","standata","standgub","standmps","stocfor1","stocfor2","tuff",
              "vtp.base","wood1p","woodw"]

type_inst_list = [('not equal by permutation','non-equivalent instances'),
                  ('equal by permutation and scaling','equivalent scaled instances'),
                  ('not equal by permutation and scaling','non-equivalent scaled instances')]

def graph_time_coordinates_solver(type_inst,name_inst_list,color,mark,solver):
        coordinates = ""
        for name_inst in name_inst_list:
            coord =  stats_solvers(type_inst,name_inst,solver)
            if coord == False:
                pass
            else:
               (size,time_second,solver) = stats_solvers(type_inst,name_inst,solver)
               coordinates = coordinates + "("+ str(size) + "," + str(time_second) +")\n"
        coordonate = add_plot(color,mark,coordinates)
        file.write(coordonate)


def graphic_time_solvers(type_inst,type_name,p):
        title = "Solvers time per size of "+type_name
        graphic_head = seconds_head_env_graphi(title,type_axis,xlabel,ylabel,p)
        file.write(graphic_head)
        file.write(bissector)
        for (solver,color,mark) in solvers_:
            graph_time_coordinates_solver(type_inst,name_inst_list,color,mark,solver)
        end_env = end_env_graphic(type_axis)
        file.write(end_env)

with open("graphics/graphics.tex", "w") as file: 
    file.write(preambule1)
    file.write(preambule2)
    file.write(preambule3)
    file.write(preambule4)
    file.write(preambule5)
    list_of_graphs_legend = write_legend(solvers_strings)
    #print("list_of_graphs_legend=",list_of_graphs_legend)
    title = "Solvers time per size of equivalent instances"
    type_axis = "loglogaxis"
    first_head = first_head_env_graphi(title,type_axis,xlabel,ylabel,list_of_graphs_legend)
    file.write(first_head)
    file.write(bissector)
    for (solver,color,mark) in solvers_:
            graph_time_coordinates_solver('equal by permutation',name_inst_list,color,mark,solver)
    end_env = end_env_graphic(type_axis)
    file.write(end_env+"&")
    i = 1
    for (type_inst,type_name) in type_inst_list:
        graphic_time_solvers(type_inst,type_name,i)
        if i == 1 :
            file.write("\\\\\n")
        if i == 2 :
            file.write(" & ")
        i = i + 1
    file.write('\n')
    file.write(end_graphic)
    file.write("\\end{document}")
```
