```python
preambule1 = "\\documentclass{book}\n\\usepackage[landscape,a3paper,left=1in,right=1in,top=1in,"
preambule2 = "bottom=1in]{geometry}\n\\usepackage{amsmath}\n\\usepackage{xcolor}\n\\usepackage{tikz}\n"
preambule3 = "\\usepackage{pgfplots}\n\\usepackage{booktabs}\n\\usepackage{multirow}\n\\usepackage{amsmath,"
preambule4 = "amssymb}\n\n\\begin{document}\n\n\\vspace{2cm}\n\n"
preambule5 = "\\begin{figure}\\pgfplotsset{small,samples=10}\n\\centering\n\\begin{tabular}{lr}"

end_graphic = "\n\\end{tabular}\\\\\n\\ref{named}\\end{figure}\n\n"


def first_head_env_graphi(title,type_axis,xlabel,ylabel,list_of_graphs_legend):
    return "\\begin{tikzpicture}\n\\begin{"+type_axis+"}[scaled ticks=false, legend columns = 4,legend entries=" + list_of_graphs_legend+", legend to name = named,\ntitle={" + title + "},xlabel=" + xlabel +",ylabel=" + ylabel+"]\n"

def seconds_head_env_graphi(title,type_axis,xlabel,ylabel,p):
    if 2 == p:
        ylabel_written = ",ylabel=" + ylabel
    else:
        ylabel_written =""
        
    return "\\begin{tikzpicture}\n\\begin{"+type_axis+"}[scaled ticks=false,\ntitle={" + title + "},xlabel=" + xlabel+ylabel_written+"]\n"

def add_function(function,min_val,max_val):
    return "\\addplot[domain="+min_val+":"+max_val+"] {"+ function + "};\n"
    
def add_plot(color,mark,coordinates):
    return "\\addplot[only marks,color="+color+",mark="+mark+"] coordinates {"+ coordinates + "};\n"   
 
```

```python

def end_env_graphic(type_axis):
    return "\\end{"+type_axis+"}\n\\end{tikzpicture}\n"

models = [('permutation_model','black','oplus'),('quadratic_model','blue','otimes'),('theorem_model','red','o'),('algo_test','teal','*'),('enhanced_test','brown','diamond')]
models_strings = ['\\mbox{  }permutation\\_model\\mbox{  }','\\mbox{  }quadratic\\_model\\mbox{  }','\\mbox{  }theorem\\_model\\mbox{  }','\\mbox{  }algo\\_test\\mbox{  }','\\mbox{  }enhanced\\_test']
solvers = ['coinbc','cp-sat','chuffed','gecode','algo_test','enhanced_test','sicstus']
type_inst_list = ['equal by permutation','not equal by permutation']
results_test_list = ['equivalent','not equivalent','unknown due to timeout']
ylabel = "Time to solve an instance (min)"
xlabel = "Number of constraints + variables"
```

```python
import datetime

def compute_time(time_string):
    format_seconds = "%H:%M:%S"
    date_time = datetime.datetime.strptime(time_string, format_seconds)
    time_delta = datetime.timedelta(hours=date_time.hour,minutes=date_time.minute,seconds=date_time.second)
    return time_delta.total_seconds()//60

def compute_coord_time(instance_name,type_inst,solver,model):
    portion = 0
    df_time = df[(df['solver'] == solver) & (df['type of instance'] == type_inst) & (df['instance name'] == instance_name) & (df['model'] == model)]
    if df_time.empty:
        return 140
    else :
        if type_inst == 'not equal by permutation' and model =='algo_test':
            #print(df_time['execution time'])
            pass
        for time_string in df_time['execution time']:
            return compute_time(time_string)
       
    
def graph_coordinates_success_time(type_inst,model,solver,color,mark,file):
    coordinates = ""
    if model == 'enhanced_test' and (type_inst == 'equal by permutation' or type_inst == 'not equal by permutation'):
            coordinates = "(100,140)\n"
    
    
    else :
        for instance_name in name_inst_list:
            df_name = df[df['instance name'] == instance_name]['Unnamed: 0']
            if df_name.empty :
                pass
            else :
                ind = df[df['instance name'] == instance_name]['Unnamed: 0'].tolist()[0]
                x = str(df.loc[ind,'number of variables'] + df.loc[ind,'number of constraints'])
                y0 = compute_coord_time(instance_name,type_inst,solver,model)
                if y0 < 1:
                    y = '1'
                else :
                    y = str(compute_coord_time(instance_name,type_inst,solver,model))
                coordinates = coordinates + "("+ x + "," + y +")\n"
    coordonate = add_plot(color,mark,coordinates)
    file.write(coordonate)
```

```python
models = {'permutation_model':'black','quadratic_model':'blue','theorem_model':'teal','algo_test':'brown','enhanced_test':'red'}
solvers = {'coinbc':'pentagon','cp-sat':'star','chuffed':'triangle','gecode':'o','cplex':'square','scip':'diamond','algo_test':'oplus','enhanced_test':'otimes'}

models_ = {'permutation\\_model':'black','quadratic\\_model':'blue','theorem\\_model':'teal','algo\\_test':'brown','enhanced\\_test':'red'}
solvers_ = {'coinbc':'pentagon','cp-sat':'star','chuffed':'triangle','gecode':'o','cplex':'square','scip':'diamond','algo\\_test':'oplus','enhanced\\_test':'otimes'}


type_inst_list = ['not equal by permutation','equal by permutation and scaling',
                  'not equal by permutation and scaling']

def color_mark(models,solvers,algo_enhanced):
    avail_res = {}
    for model in models.keys():
        for solver in solvers.keys():
            if model not in algo_enhanced and solver not in algo_enhanced:
                avail_res[(model,solver)]= [models[model],solvers[solver]]
    avail_res[(algo_enhanced[0],algo_enhanced[0])] = ['brown','oplus']
    avail_res[(algo_enhanced[1],algo_enhanced[1])] = ['red','otimes']
    return avail_res

def write_legend(models_,solvers_):
    i = 0
    legend ="{"
    available_res = color_mark(models_,solvers_,['algo\\_test','enhanced\\_test'])
    for (model,solver) in available_res.keys():
        name = model + "\\_" + solver
        if i == 0:
               legend = legend + name
        elif i == len(available_res)-1:
                legend = legend + ";," + name + "}"
        else:
                legend = legend + ";," + name
        i = i + 1
    return legend

def gen_coordinate(type_inst,available_res,file):
    for (model,solver) in available_res.keys():
            color = available_res[(model,solver)][0]
            mark = available_res[(model,solver)][1]
            graph_coordinates_success_time(type_inst,model,solver,color,mark,file)
        

def graphic_model(type_inst,available_res,p,file):
        title = "Solving size for "+type_inst
        graphic_head = seconds_head_env_graphi(title,type_axis,xlabel,ylabel,p)
        file.write(graphic_head)
        gen_coordinate(type_inst,available_res,file)
        end_env = end_env_graphic(type_axis)
        file.write(end_env)
```

```python
def gen_graph_size(models,solvers,models_string,solvers_string,directory):
    with open("graphics/"+directory+"/graphics.tex", "w") as file: 
        file.write(preambule1)
        file.write(preambule2)
        file.write(preambule3)
        file.write(preambule4)
        file.write(preambule5)
        list_of_graphs_legend = write_legend(models_string,solvers_string)
        title = "Solving size for equal by permutation"
        type_axis = "loglogaxis"
        first_head = first_head_env_graphi(title,type_axis,xlabel,ylabel,list_of_graphs_legend)
        file.write(first_head)
        available_res = color_mark(models,solvers,['algo_test','enhanced_test'])
        gen_coordinate('equal by permutation',available_res,file)
        end_env = end_env_graphic(type_axis)
        file.write(end_env+"&")
        i = 1
        for type_inst in type_inst_list:
            graphic_model(type_inst,available_res,i,file)
            if i == 1 :
                file.write("\\\\\n")
            if i == 2 :
                file.write(" & ")
            i = i + 1
        file.write('\n')
        file.write(end_graphic)
        file.write("\\end{document}")
```
